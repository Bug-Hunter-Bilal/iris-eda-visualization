# Visualization

## Scatter plots

```python
plt.figure(figsize=(8,5))
sns.scatterplot(x = 'SepalLengthCm', y = 'SepalWidthCm', hue = 'Species', data = df )
plt.title("Sepal Length vs Sepal Width")
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.scatterplot(x='PetalLengthCm', y='PetalWidthCm', hue = 'Species', data = df)
plt.show()
```

## Histograms

```python
plt.figure(figsize=(8,5))
sns.histplot(df['SepalLengthCm'], kde = True)
plt.show()
```

```python
df.columns
```

```python
plt.figure(figsize=(8,5))
sns.histplot(df['SepalWidthCm'], kde = True)
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.histplot(df['PetalLengthCm'], kde = True)
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.histplot(df['PetalWidthCm'], kde = True)
plt.show()
```

## Box Plots

```python
plt.figure(figsize=(8,5))
sns.boxplot(y = df['SepalLengthCm'])
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.boxplot(x = df['SepalWidthCm'])
plt.show()
```

```python
df.columns
```

```python
plt.figure(figsize=(8,5))
sns.boxplot(x = 'Species', y = 'SepalLengthCm', data = df)
plt.show()
```

## Heatmap

```python
df.drop('Id', axis=1, inplace = True)
```

```python
correaltion = df.corr(numeric_only = True)
```

```python
plt.figure(figsize=(8,5))
sns.heatmap(correaltion, square = True, annot = True, fmt = '.1f', annot_kws = {'size':12})
```
